```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.neural_network import MLPRegressor

# Import data
data_df = pd.read_csv("Project2_data_sypder.csv")


# Convert data to array to maybe avoid error
data_df_array = data_df.to_numpy()

#Seperating Data into the features(X) and Targets(Y)
X = data_df_array[:, 1:43]
Y = data_df_array[:, 43:]


# Scaling only the X (not necessary for Y)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Splitting data into test and train for each X and Y
TrainingX, TestingX, TrainingY, TestingY = train_test_split(X_scaled, Y, test_size=0.33, random_state=42)

# Parameters for grid search
param_space = [{
    'hidden_layer_sizes': [80],
    'activation': ['relu'],
    'solver': ['adam'],
    'learning_rate': ['adaptive']
}]

# Grid search with MLPRegressor
regr = GridSearchCV(MLPRegressor(max_iter=30000, random_state=1), param_space)
regr.fit(TrainingX, TrainingY)


#put pickle here
import pickle

#save model to the disk
filename = 'final_model.sav'
pickle.dump(regr, open(filename, 'wb'))


Ypredictions = regr.predict(TestingX)


plt.scatter(TestingY[:,14], Ypredictions[:,14])
plt.show()
```
